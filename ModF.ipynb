The derivative is a very important concept in the world of computing.  Essentially, the derivative of a function is a new function that gives the rate of change of the original function at every point.  One way to determine the derivative is by using the rules of calculus, such as the power rule, product rule, chain rule, and the rest.  But In real life, not all functions are so easily defined mathematically.  In these cases, you have to find the derivative by other means.

A numerical grid is a series of numbers that are evenly spaced.  This spacing is called the step size.  Functions in real life are usually continuous, but when you implement them with python, they are often discretized.  This is done with a numerical grid.  In python, we use linspace.

The slope between 2 points is easy to compute.  The slope at a single point is actually defined as the limit of the slope between 2 points as they get closer together.  When a function is discretized, you can not get 2 points arbitrarily close together.  Instead, to calculate the slope at 1 point, you can just use the slope between the current point (x_1,f(x_i)) and the next one in the numerical grid (x_i+1, f(x_i+1)).  This is called the forward difference.  There is also the backward difference, where you use the previous one in the numerical grid.  The central difference is when you take the slope between the next point and the previous point.

Another approach is the Taylor series.  A Taylor series is a polynomial approximation of any function.  You pick 1 point and then approximate the function by making sure the approximation has the same derivative, second derivative, third, fourth, fifth, and so on as the original function at that point.

Now we will use python to demonstrate the forward difference approximation.

```python
import numpy as np
import matplotlib.pyplot as plt

# step size
h = 0.5

# numerical grid for x and f(x) = sin(x)
x = np.arange(0, 6.3, h)
y = np.sin(x)

# the real derivative is cos(x), but diff is the forward approximation
diff = np.diff(y)/h
deriv = np.cos(x)

plt.plot(x, deriv, "c")
plt.plot(x[:-1:], diff, "g")
plt.legend(["actual derivative", "forward approximation"])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# using a smaller step size gives a better approximation
h = 0.1

# numerical grid for x and f(x) = sin(x)
x = np.arange(0, 6.3, h)
y = np.sin(x)

# the real derivative is cos(x), but diff is the forward approximation
diff = np.diff(y)/h
deriv = np.cos(x)

plt.plot(x, deriv, "c")
plt.plot(x[:-1:], diff, "g")
plt.legend(["actual derivative", "forward approximation"])
```

Most real life functions will not be as smooth as a simple sine wave.  They will very likely have noise (unpredictable fluctuations).  This can have a disproportionately large effect on derivatives.  Choosing a step size must be carefully considered.  Sometimes, as step size gets smaller, error can actually increase.

```python
import numpy as np
import matplotlib.pyplot as plt

# step size
h1 = 0.01
h2 = 0.03
h3 = 0.1
h4 = 0.3

def f(x):
  return np.sin(x) + 0.1*np.sin(11*x)  - 0.07*np.cos(23.7*x - 2) + 0.08*np.sin(30*x+7)

# numerical grid for x and f(x) = sin(x)
x1 = np.arange(0, 6.3, h1)
x2 = np.arange(0, 6.3, h2)
x3 = np.arange(0, 6.3, h3)
x4 = np.arange(0, 6.3, h4)
y1 = f(x1)
y2 = f(x2)
y3 = f(x3)
y4 = f(x4)

# the real derivative is cos(x), but diff is the forward approximation
d1 = np.diff(y1)/h1
d2 = np.diff(y2)/h2
d3 = np.diff(y3)/h3
d4 = np.diff(y4)/h4

#plt.plot(x, deriv, "c")
plt.plot(x1, y1, "c")
#plt.plot(x1[:-1:], d1, "r")
plt.plot(x2[:-1:], d2, "y--")
#plt.plot(x3[:-1:], d3, "g-o")
plt.plot(x4[:-1:], d4, "b")
plt.legend(["f(x)", "f'(x), h = 0.03", "f'(x), h = 0.3"])
```
